```python
#INGÓSÁG RAW DATA

# Define a dataframe to store your structured data

vny_2025_jarmu = pd.DataFrame(columns = ['oldal', 'sor', 'PID', 'name', 'ingóság száma', 'jármű fajtája', 'járművek száma', 'jármű típus száma', 'Gépjárművek', 'szerz. ideje, jogcíme', 'which if'])

# to track which lines to write in
counter = 0

# Person ID - to track individuals
pid = int(0)

#for i in range(len(reader.pages)):
for i in range(600):
    page = reader.pages[i]
    text = page.extract_text()
    t_lines = list(text.split('\n'))
    
    # to reach back and forward on the page if needed
    linecount = 0
    
    for line in t_lines:
        linecount += 1
        if '2. A nyilatkozatot adó neve:' in line:
            counter += 1
            pid += 1
            ingo_sz = 0
            jarmu = ''
            jarmu_sz = 0
            
            name = str(line.removeprefix('2. A nyilatkozatot adó neve: ').strip())
            

        if 'a) személygépkocsi:' in line and not '----- típus' in line:
            if not '"-"' in line.strip():
                counter += 1 if not counter in vny_2025_jarmu.index or vny_2025_jarmu['name'][counter] == name else 0
                which_if = 'a'
                ingo_sz += 1
                jarmu = 'SZGK'
                jarmu_sz += 1
                typ_sz = 1
                ingo = str(line.removeprefix('a) személygépkocsi:').strip())
                ingo = ingo.removesuffix('típus').strip()
                id_jogc = t_lines[linecount].strip().removeprefix('a szerzés ideje, jogcíme:').strip()
                
                vny_2025_jarmu.loc[counter, 'name'] = name
                vny_2025_jarmu.loc[counter, 'PID'] = int(pid)
                vny_2025_jarmu.loc[counter, 'Gépjárművek'] = ingo
                vny_2025_jarmu.loc[counter, 'ingóság száma'] = ingo_sz
                vny_2025_jarmu.loc[counter, 'jármű fajtája'] = jarmu
                vny_2025_jarmu.loc[counter, 'járművek száma'] = jarmu_sz
                vny_2025_jarmu.loc[counter, 'jármű típus száma'] = typ_sz
                vny_2025_jarmu.loc[counter, 'szerz. ideje, jogcíme'] = id_jogc
                vny_2025_jarmu.loc[counter, 'oldal'] = int(i+1)
                vny_2025_jarmu.loc[counter, 'sor'] = int(linecount)
                vny_2025_jarmu.loc[counter, 'which if'] = which_if
                continue

        if 'b) tehergépjármű' in line and not '----- típus' in line:
            if not '"-"' in line.strip():
                counter += 1 if not counter in vny_2025_jarmu.index or vny_2025_jarmu['name'][counter] == name else 0
                which_if = 'b'
                ingo_sz += 1
                jarmu = 'TEHER-GK'
                jarmu_sz += 1
                typ_sz = 1
                ingo = str(line.removeprefix('b) tehergépjármű, autóbusz:').strip())
                ingo = ingo.removesuffix('típus').strip()
                id_jogc = t_lines[linecount].strip().removeprefix('a szerzés ideje, jogcíme:').strip()
                
                vny_2025_jarmu.loc[counter, 'name'] = name
                vny_2025_jarmu.loc[counter, 'PID'] = int(pid)
                vny_2025_jarmu.loc[counter, 'Gépjárművek'] = ingo
                vny_2025_jarmu.loc[counter, 'ingóság száma'] = ingo_sz
                vny_2025_jarmu.loc[counter, 'jármű fajtája'] = jarmu
                vny_2025_jarmu.loc[counter, 'járművek száma'] = jarmu_sz
                vny_2025_jarmu.loc[counter, 'jármű típus száma'] = typ_sz
                vny_2025_jarmu.loc[counter, 'szerz. ideje, jogcíme'] = id_jogc
                vny_2025_jarmu.loc[counter, 'oldal'] = int(i+1)
                vny_2025_jarmu.loc[counter, 'sor'] = int(linecount)
                vny_2025_jarmu.loc[counter, 'which if'] = which_if
                continue

        if 'c) motorkerékpár' in line and not '----- típus' in line:
            if not '"-"' in line.strip():
                counter += 1 if not counter in vny_2025_jarmu.index or vny_2025_jarmu['name'][counter] == name else 0
                which_if = 'c'
                ingo_sz += 1
                jarmu = 'MOTOR'
                jarmu_sz += 1
                typ_sz = 1
                ingo = str(line.removeprefix('c) motorkerékpár:').strip())
                ingo = ingo.removesuffix('típus').strip()
                id_jogc = t_lines[linecount].strip().removeprefix('a szerzés ideje, jogcíme:').strip()
                
                vny_2025_jarmu.loc[counter, 'name'] = name
                vny_2025_jarmu.loc[counter, 'PID'] = int(pid)
                vny_2025_jarmu.loc[counter, 'Gépjárművek'] = ingo
                vny_2025_jarmu.loc[counter, 'ingóság száma'] = ingo_sz
                vny_2025_jarmu.loc[counter, 'jármű fajtája'] = jarmu
                vny_2025_jarmu.loc[counter, 'járművek száma'] = jarmu_sz
                vny_2025_jarmu.loc[counter, 'jármű típus száma'] = typ_sz
                vny_2025_jarmu.loc[counter, 'szerz. ideje, jogcíme'] = id_jogc
                vny_2025_jarmu.loc[counter, 'oldal'] = int(i+1)
                vny_2025_jarmu.loc[counter, 'sor'] = int(linecount)
                vny_2025_jarmu.loc[counter, 'which if'] = which_if
                continue

        
        if '2. Vízi vagy légi jármű:' in line and not '-----' in t_lines[linecount] and not '"-"' in t_lines[linecount].strip():
            
            counter += 1 if not counter in vny_2025_jarmu.index or vny_2025_jarmu['name'][counter] == name else 0
            which_if = 'vizi'
            ingo_sz += 1
            jarmu = 'VÍZI-LÉGI'
            jarmu_sz += 1
            typ_sz = 1
            ingo_1 = t_lines[linecount].removeprefix('jellege:').strip()
            ingo_2 = t_lines[linecount+1].strip().removeprefix('típusa:').strip()
            ingo = f'{ingo_1}, {ingo_2}'
            id_jogc = t_lines[linecount+2].strip().removeprefix('a szerzés ideje, jogcíme:').strip()
            
            vny_2025_jarmu.loc[counter, 'name'] = name
            vny_2025_jarmu.loc[counter, 'PID'] = int(pid)
            vny_2025_jarmu.loc[counter, 'Gépjárművek'] = ingo
            vny_2025_jarmu.loc[counter, 'ingóság száma'] = ingo_sz
            vny_2025_jarmu.loc[counter, 'jármű fajtája'] = jarmu
            vny_2025_jarmu.loc[counter, 'járművek száma'] = jarmu_sz
            vny_2025_jarmu.loc[counter, 'jármű típus száma'] = typ_sz
            vny_2025_jarmu.loc[counter, 'szerz. ideje, jogcíme'] = id_jogc
            vny_2025_jarmu.loc[counter, 'oldal'] = int(i+1)
            vny_2025_jarmu.loc[counter, 'sor'] = int(linecount)
            vny_2025_jarmu.loc[counter, 'which if'] = which_if
            continue

        if 'típus' in line and not '----- típus' in line and jarmu == 'SZGK' and jarmu_sz > 0 and not '"-"' in line.strip():
            if not 'típusa: -----' in line.strip():
                which_if = 'a-2'
                counter += 1
                ingo_sz += 1
                jarmu_sz += 1
                typ_sz += 1
                ingo = str(line.removesuffix('típus').strip())
                id_jogc = t_lines[linecount].strip().removeprefix('a szerzés ideje, jogcíme:').strip()
    
                vny_2025_jarmu.loc[counter, 'name'] = name
                vny_2025_jarmu.loc[counter, 'PID'] = int(pid)
                vny_2025_jarmu.loc[counter, 'Gépjárművek'] = ingo
                vny_2025_jarmu.loc[counter, 'ingóság száma'] = ingo_sz
                vny_2025_jarmu.loc[counter, 'jármű fajtája'] = jarmu
                vny_2025_jarmu.loc[counter, 'járművek száma'] = jarmu_sz
                vny_2025_jarmu.loc[counter, 'jármű típus száma'] = typ_sz
                vny_2025_jarmu.loc[counter, 'szerz. ideje, jogcíme'] = id_jogc
                vny_2025_jarmu.loc[counter, 'oldal'] = int(i+1)
                vny_2025_jarmu.loc[counter, 'sor'] = int(linecount)
                vny_2025_jarmu.loc[counter, 'which if'] = which_if
                continue



        if 'típus' in line and not '----- típus' in line and jarmu == 'TEHER-GK' and jarmu_sz > 0 and not '"-"' in line.strip():
            if not 'típusa: -----' in line.strip():
                which_if = 'b2'
                counter += 1
                ingo_sz += 1
                jarmu_sz += 1
                typ_sz += 1
                ingo = str(line.removesuffix('típus').strip())
                id_jogc = t_lines[linecount].strip().removeprefix('a szerzés ideje, jogcíme:').strip()
    
                vny_2025_jarmu.loc[counter, 'name'] = name
                vny_2025_jarmu.loc[counter, 'PID'] = int(pid)
                vny_2025_jarmu.loc[counter, 'Gépjárművek'] = ingo
                vny_2025_jarmu.loc[counter, 'ingóság száma'] = ingo_sz
                vny_2025_jarmu.loc[counter, 'jármű fajtája'] = jarmu
                vny_2025_jarmu.loc[counter, 'járművek száma'] = jarmu_sz
                vny_2025_jarmu.loc[counter, 'jármű típus száma'] = typ_sz
                vny_2025_jarmu.loc[counter, 'szerz. ideje, jogcíme'] = id_jogc
                vny_2025_jarmu.loc[counter, 'oldal'] = int(i+1)
                vny_2025_jarmu.loc[counter, 'sor'] = int(linecount)
                vny_2025_jarmu.loc[counter, 'which if'] = which_if
                continue

        if 'típus' in line and not '----- típus' in line and jarmu == 'MOTOR' and jarmu_sz > 0 and not '"-"' in line.strip():
            if not 'típusa: -----' in line.strip():
                which_if = 'c2'
                counter += 1
                ingo_sz += 1
                jarmu_sz += 1
                typ_sz += 1
                ingo = str(line.removesuffix('típus').strip())
                id_jogc = t_lines[linecount].strip().removeprefix('a szerzés ideje, jogcíme:').strip()
    
                vny_2025_jarmu.loc[counter, 'name'] = name
                vny_2025_jarmu.loc[counter, 'PID'] = int(pid)
                vny_2025_jarmu.loc[counter, 'Gépjárművek'] = ingo
                vny_2025_jarmu.loc[counter, 'ingóság száma'] = ingo_sz
                vny_2025_jarmu.loc[counter, 'jármű fajtája'] = jarmu
                vny_2025_jarmu.loc[counter, 'járművek száma'] = jarmu_sz
                vny_2025_jarmu.loc[counter, 'jármű típus száma'] = typ_sz
                vny_2025_jarmu.loc[counter, 'szerz. ideje, jogcíme'] = id_jogc
                vny_2025_jarmu.loc[counter, 'oldal'] = int(i+1)
                vny_2025_jarmu.loc[counter, 'sor'] = int(linecount)
                vny_2025_jarmu.loc[counter, 'which if'] = which_if
                continue

```
